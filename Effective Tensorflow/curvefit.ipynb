```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,200)
y = 5*np.power(x,2) + 3
ns = np.random.normal(-10,10,200)
yn = y + ns
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,yn)
```

```python
tx = tf.placeholder(tf.float64)
ylab = tf.placeholder(tf.float64)

with tf.variable_scope("w",reuse=tf.AUTO_REUSE ): 
    w0 = tf.get_variable(name='w0',shape=(1,1),dtype=tf.float64)
    w1 = tf.get_variable(name='w1',shape=(1,1),dtype=tf.float64)
    w2 = tf.get_variable(name='w2',shape=(1,1),dtype=tf.float64)
ty = w0*tf.square(tx) + w1*tx + w2

# ylab = tf.constant(yn)
# loss = tf.nn.l2_loss(ty-ylab)
loss = tf.reduce_mean((ty-ylab)**2)
with tf.variable_scope("w",reuse=tf.AUTO_REUSE ): 
    train_op = tf.train.AdamOptimizer(0.5).minimize(loss)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for _ in range(1000):
    _,ls = sess.run([train_op,loss],feed_dict={tx:x,ylab:y})
    print(ls)
```
